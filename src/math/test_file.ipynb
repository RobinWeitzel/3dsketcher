```python
import numpy as np

ih = np.array([1, 0, 0])
jh = np.array([0, 1, 0])
kh = np.array([0, 0, 1])

Rx = lambda phi: np.array([[0, 0, 0],
                           [0, np.cos(phi), -np.sin(phi)],
                           [0, np.sin(phi), np.cos(phi)]])
Ry = lambda phi: np.array([[np.cos(phi), 0, np.sin(phi)],
                           [0, 1, 0],
                           [-np.sin(phi), 0, np.cos(phi)]])
Rz = lambda phi: np.array([[np.cos(phi), -np.sin(phi), 0],
                           [np.sin(phi), np.cos(phi), 0],
                           [0, 0, 1]])
```

```python
phi = -0.25*np.pi
Ap = 1/np.diag(Ry(phi)) * Ry(phi)

v = np.array([1,1,0])
Ry(-phi) @ Ap @ v
```
